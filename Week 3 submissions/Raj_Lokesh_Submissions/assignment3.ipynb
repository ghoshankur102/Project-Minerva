```python
import os
import re
from typing import TypedDict, List
from google.colab import userdata
from langgraph.graph import StateGraph, END
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python

# Authentication keys setup
os.environ["GOOGLE_API_KEY"] = userdata.get("GOOGLE_API_KEY")

# Deep-learning engine declaration
inference_core = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite",
    temperature=0.2
)
```

```python
# %%
class PuzzleContext(TypedDict):
    """The runtime data container tracking optimization progress across nodes."""
    raw_challenge_text: str
    architectural_plan: str
    candidate_source_code: str
    audit_verdict: str
    loop_index: int
    error_registry: List[str]
    synthetic_test_matrix: str
```

```python
def create_simulation_vectors(state_data: PuzzleContext):
    """Generates sample diagnostic vectors to evaluate edge limits."""

    engineered_prompt = f"""
Identify three critical validation boundaries and construct isolated test frameworks for the problem below.

Target Specifications:
{state_data['raw_challenge_text']}

Required Output Paradigm:
---
[CASE 1]
Data Input Vector:
...
Expected Numerical Return:
...

[CASE 2]
Data Input Vector:
...
Expected Numerical Return:
...

[CASE 3]
Data Input Vector:
...
Expected Numerical Return:
...
---
"""
    model_output = inference_core.invoke(engineered_prompt)
    state_data["synthetic_test_matrix"] = model_output.content
    return state_data



def run_critique_engine(state_data: PuzzleContext):
    """Deep inspection of syntax parameters and dynamic programming safety."""

    engineered_prompt = f"""
Audit the provided source implementation against logic faults, memory leaks, and bounds.

Target Specifications:
{state_data['raw_challenge_text']}

Candidate Implementation:
{state_data['candidate_source_code']}

Evaluation Matrix:
{state_data['synthetic_test_matrix']}

Decision Logic Requirements:
If the source meets all constraints, output exactly: APPROVED
If any errors exist, output exactly: REJECTED - <detailed explanation>
"""
    model_output = inference_core.invoke(engineered_prompt)
    state_data["audit_verdict"] = model_output.content.strip()
    return state_data


def diagnose_logic(state_data: PuzzleContext):
    """Extracts conceptual schemas and time complexity thresholds."""

    engineered_prompt = f"""
Deconstruct the optimization problem listed below. Synthesize a core methodology outline.

Target Specifications:
{state_data['raw_challenge_text']}

Ensure the technical blueprint targets:
A) Conceptual Core Strategy
B) Algorithmic Categorization
C) Upper Bound Big-O Complexity
D) Edge Case Constraints

Provide a clean, dense overview.
"""
    model_output = inference_core.invoke(engineered_prompt)
    state_data["architectural_plan"] = model_output.content
    return state_data


def compile_production_source(state_data: PuzzleContext):
    """Generates standardized production C++ logic maps targeting standard libraries."""

    historical_anomalies = "\n".join(state_data["error_registry"])

    engineered_prompt = f"""
Develop an optimized C++17 solution block to successfully compute the challenge framework.

Target Specifications:
{state_data['raw_challenge_text']}

Structural Blueprint:
{state_data['architectural_plan']}

Avoid Repeating These Documented Failures:
{historical_anomalies}

Do not include markdown tags or explanation blocks. Provide only pure valid C++ code strings.
"""
    model_output = inference_core.invoke(engineered_prompt)
    state_data["candidate_source_code"] = model_output.content
    return state_data


def record_exception_state(state_data: PuzzleContext):
    """Appends debugging metadata into long-term history states."""

    state_data["error_registry"].append(state_data["audit_verdict"])
    state_data["loop_index"] += 1
    return state_data


def evaluation_router(state_data: PuzzleContext):
    """Conditional controller routing processing flows based on verification states."""

    # Leverages regex search boundaries rather than basic substring index checks
    if re.search(r'^APPROVED', state_data["audit_verdict"], re.IGNORECASE):
        return END

    if state_data["loop_index"] >= 3:
        return END

    return "record_exception_state"
```

```python
# Pipeline Architecture Setup

graph_pipeline = StateGraph(PuzzleContext)

# Node registration phase
graph_pipeline.add_node("diagnose_logic", diagnose_logic)
graph_pipeline.add_node("compile_production_source", compile_production_source)
graph_pipeline.add_node("create_simulation_vectors", create_simulation_vectors)
graph_pipeline.add_node("run_critique_engine", run_critique_engine)
graph_pipeline.add_node("record_exception_state", record_exception_state)

# Linear flow configurations
graph_pipeline.set_entry_point("diagnose_logic")
graph_pipeline.add_edge("diagnose_logic", "compile_production_source")
graph_pipeline.add_edge("compile_production_source", "create_simulation_vectors")
graph_pipeline.add_edge("create_simulation_vectors", "run_critique_engine")

# Conditional dependency branch
graph_pipeline.add_conditional_edges(
    "run_critique_engine",
    evaluation_router
)
graph_pipeline.add_edge(
    "record_exception_state",
    "compile_production_source"
)

# Compile pipeline logic
runtime_agent = graph_pipeline.compile()
```

```python
source_challenge = """
You are given n points on a circle. Count the number of ways to choose exactly 3 points
such that the triangle formed contains the center of the circle.
Answer modulo 10^9+7.

Input:
First line: integer n (3 ≤ n ≤ 10^5)
Second line: n integers — angles of the points in degrees (0 ≤ a_i < 360, all distinct)

Output:
One integer — the number of valid triangles mod 10^9+7.
"""

# %%
# Parameter mapping instantiation
initialization_vector = {
    "raw_challenge_text": source_challenge,
    "architectural_plan": "",
    "candidate_source_code": "",
    "audit_verdict": "",
    "loop_index": 0,
    "error_registry": [],
    "synthetic_test_matrix": ""
}

# Run orchestration workflow
runtime_results = runtime_agent.invoke(initialization_vector)

# %%
print("\n" + "=" * 60)
print("             EXECUTION PROCESS TERMINATION STATUS            ")
print("=" * 60)

print("\n[CONTEXT DETAILED PARAMETERS]")
print("~" * 40)
print(runtime_results["raw_challenge_text"])

print("\n[ALGORITHMIC SCHEDULING PLAN]")
print("~" * 40)
print(runtime_results["architectural_plan"])

print("\n[ISOLATED SIMULATION MATRIX]")
print("~" * 40)
print(runtime_results["synthetic_test_matrix"])

print("\n[AUTOMATED AUDIT DECISION]")
print("~" * 40)
print(runtime_results["audit_verdict"])

print("\n[TRACKING COEFFICIENTS]")
print("~" * 40)
print(f"Cycle Iteration Bounds Used: {runtime_results['loop_index']}")
print(f"Total Registry Entries Logged: {len(runtime_results['error_registry'])}")

print("\n[COMPILED SOURCE STRINGS]")
print("~" * 40)
print(runtime_results["candidate_source_code"])

print("\n" + "=" * 60)
print("             METRICS PROCESSOR SYSTEM EXITED SUCCESS         ")
print("=" * 60)
```
