```python
from sentence_transformers import SentenceTransformer # Focuses on meaning and context of the sentences. It is designed to capture the overall meaning of the sentences, including the relationships between words and the context in which they are used.
model = SentenceTransformer('all-MiniLM-L6-v2') # the model is defined here. This model is already trained on huge amounts of text.

# model becomes an object that can do things like encode sentences, compute embeddings, compare similarities

# Sentences to be encoded
sentences = [
    "Virat Kohli is a great cricketer.",
    "Sachin Tendulkar is a great cricketer."
]
embeddings = model.encode(sentences)
```

```python
embeddings.shape # The shape of the embeddings is (2, 384) because we have 2 sentences and each sentence is represented by a 384-dimensional vector.
```

```python
embeddings
```

```python
similarity=model.similarity(embeddings,embeddings)
similarity # depicts the similarity between the two sentences. The value will be between 0 and 1, where 1 means the sentences are identical and 0 means they are completely different.
```

```python
from sentence_transformers import CrossEncoder
model = CrossEncoder('cross-encoder/stsb-roberta-base') # this model is trained to compute the similarity between two sentences. It takes two sentences as input and outputs a similarity score between 0 and 1.

# We want to compute the similarity between the query sentence...
query = "A man is eating pasta."

# ... and all sentences in the corpus
corpus = [
    "A man is eating food.",
    "A man is eating a piece of bread.",
    "The girl is carrying a baby.",
    "A man is riding a horse.",
    "A woman is playing violin.",
    "Two men pushed carts through the woods.",
    "A man is riding a white horse on an enclosed ground.",
    "A monkey is playing drums.",
    "A cheetah is running behind its prey.",
]

ranks=model.rank(query, corpus) # this will compute the similarity between the query and each sentence in the corpus and return the ranks of the sentences in the corpus based on their similarity to the query.
```

```python
ranks # the output will be a list of indices that represent the ranks of the sentences in the corpus based on their similarity to the query. The index 0 corresponds to the most similar sentence, index 1 corresponds to the second most similar sentence, and so on.
```

```python
from sentence_transformers import SparseEncoder # Focuses on words present in the input sentences and ignores the rest. It is designed to capture the most important features of the sentences while discarding less relevant information.
model = SparseEncoder("naver/splade-cocondenser-ensembledistil") # this model is trained to compute the similarity between two sentences using sparse representations. It takes two sentences as input and outputs a similarity score between 0 and 1.
```

```python
sentences = [
    "The weather is lovely today.",
    "It's so sunny outside!",
    "He drove to the stadium.",
]

embeddings = model.encode(sentences) # this will compute the sparse embeddings for the given sentences. The output will be a list of sparse vectors, where each vector represents the corresponding sentence in the input list.
embeddings
```

```python
embeddings.shape # The shape of the embeddings will depend on the specific implementation of the SparseEncoder model.
```

```python
similarity = model.similarity(embeddings, embeddings) # this will compute the similarity between the sparse embeddings of the sentences. The output will be a matrix of similarity scores, where each entry (i, j) represents the similarity between the i-th and j-th sentences in the input list.
similarity
```
