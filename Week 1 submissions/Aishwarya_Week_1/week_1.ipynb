```python
import requests
from bs4 import BeautifulSoup

url = "https://indianexpress.com/article/explained/explained-global/russia-china-relations-putin-xi-beijing-visit-military-alliance-10699680/?ref=hometop_hp"
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

article = soup.find("div", id="pcl-full-content")
paragraphs = article.find_all("p")

text = ""
for p in paragraphs:
    text = text + p.text + "\n"

file = open("indian_express_article.txt", "w", encoding="utf-8")
file.write(text)
file.close()
```

```python
import nltk
from nltk.tokenize import sent_tokenize

nltk.download('punkt_tab')

file = open("indian_express_article.txt", "r", encoding="utf-8")
text_data = file.read()
file.close()

all_sentences = sent_tokenize(text_data)

print(all_sentences)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(all_sentences)

print(embeddings.shape)
```

```python
similarities = model.similarity(embeddings, embeddings)
print(similarities)
```
