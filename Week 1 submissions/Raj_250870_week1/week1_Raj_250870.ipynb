```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://indianexpress.com/article/political-pulse/cockroach-janta-party-bjp-sees-bid-to-destabilise-voices-within-allies-call-for-listening-to-young-10704844/'
response = requests.get(url)
response.status_code
```

```python
soup=BeautifulSoup(response.content, 'html.parser')
soup.title.text
```

```python
soup.text
```

```python
len(soup.text)
```

```python
paragraphs = soup.find_all('p')
```

```python
article = ' '
for p in paragraphs:
    text=p.get_text(strip=True)
    article += text + ' '
article[:2000]
```

```python
with open("cockroach.txt", "w", encoding="utf-8") as f:
    f.write(article)
```
