```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
```

```python
url = "https://timesofindia.indiatimes.com/city/mumbai/pillion-rider-crushed-by-best-bus-after-biker-applies-sudden-brake-to-avoid-hitting-pedestrian-crossing-andheri-road-in-mumbai/articleshow/131230658.cms"
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
    "Referer": "https://www.google.com/"
}

response = requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
headline = soup.find("h1")

if headline:
    headline = headline.get_text(strip=True)
else:
    headline = "Headline not found"

print(headline)
```

```python
headline = ""
article_body = ""

# Extract structured article content
scripts = soup.find_all("script", type="application/ld+json")

for script in scripts:
    try:
        data = json.loads(script.string)

        if isinstance(data, dict) and "articleBody" in data:
            headline = data.get("headline", "")
            article_body = data.get("articleBody", "")
            break

    except:
        continue
```

```python
corpus_df = pd.DataFrame({
    "headline": [headline],
    "raw_text_corpus": [article_body]
})


corpus_df.to_csv("raw_text_corpus.csv", index=False, encoding="utf-8")

print("Raw text corpus saved successfully!")
corpus_df
```
