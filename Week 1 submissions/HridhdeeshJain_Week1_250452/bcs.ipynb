```python
import requests
from bs4 import BeautifulSoup

website_url = "https://lite.cnn.com"
file_name = "News_articles.txt"

homepage = requests.get(website_url)
soup = BeautifulSoup(homepage.text, 'html.parser')
article_links = []
for a in soup.find_all('a'):
    link = a.get('href')  
    if link :
        full_link = website_url + link
        article_links.append(full_link)

article_links = list(set(article_links))
print(f"Found {len(article_links)} articles.\n")

with open(file_name, 'w', encoding='utf-8') as text_file:

    for link in article_links[:10]:
        print(f"Reading: {link}")

        article_page = requests.get(link)
        article_soup = BeautifulSoup(article_page.text, 'html.parser')
    
        title_tag = article_soup.find('h2')
        title = title_tag.text if title_tag else "No Title Found"

        text_file.write(title + "\n\n")

        for p in article_soup.find_all('p'):
            
            text = p.text.strip()
            if text == "":
                continue
            if "See Full Web Article" in text or "Cable News Network" in text or "Terms of Use" in text or "Privacy Policy" in text:
                continue
            text_file.write(p.text + "\n")
        text_file.write("\n" + "="*50 + "\n\n")

print("\nOpen News_articles.txt .")
```
