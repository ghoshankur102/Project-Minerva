```python
import json

with open("corpus.json", "r", encoding="utf-8") as f:
    data = json.load(f)

print(type(data))
```

```python
if isinstance(data, list):
    print("Length:", len(data))
    print(data[0])
else:
    print(data.keys())
```

```python
all_text = ""

for article in data:
    all_text += article["title"] + "\n"
    all_text += article["text"] + "\n\n"

print(all_text[:1000])
```

```python
import re

clean_text = all_text.lower()

clean_text = re.sub(r'[^\w\s]', '', clean_text)

print(clean_text[:1000])
```

```python
print("Words:", len(clean_text.split()))
```

```python
chunks = []

words = clean_text.split()

chunk_size = 300

for i in range(0, len(words), chunk_size):
    chunk = " ".join(words[i:i+chunk_size])
    chunks.append(chunk)

print("Total chunks:", len(chunks))
```

```python
from rank_bm25 import BM25Okapi

tokenized_corpus = [chunk.split() for chunk in chunks]

bm25 = BM25Okapi(tokenized_corpus)

print("BM25 Ready")
```

```python
query = "summer holidays"

scores = bm25.get_scores(query.split())

print(scores)
```

```python
import numpy as np

top_chunks = np.argsort(scores)[::-1][:3]

for idx in top_chunks:
    print(f"\nChunk {idx}")
    print(chunks[idx][:500])
    print("="*60)
```

```python
import numpy as np

def retrieve(query, k=3):
    scores = bm25.get_scores(query.split())

    top_indices = np.argsort(scores)[::-1][:k]

    results = []

    for idx in top_indices:
        results.append(chunks[idx])

    return results
```

```python
docs = retrieve("summer holidays")

print(docs[0][:500])
```

```python
def build_context(docs):
    return "\n\n".join(docs)
```

```python
def ask_rag(question):

    docs = retrieve(question)

    context = build_context(docs)

    prompt = f"""
Use ONLY the provided context.

Context:
{context}

Question:
{question}

Answer:
"""

    response = llm.invoke(prompt)

    return response.content
```

```python
answer = ask_rag(
    "What did the government announce for families during summer holidays?"
)

print(answer)
```

```python
full_corpus = clean_text
```

```python
def ask_naive(question):

    prompt = f"""
Context:
{full_corpus}

Question:
{question}

Answer:
"""

    response = llm.invoke(prompt)

    return response.content
```

```python
question = "What did the government announce for families during summer holidays?"

print("RAG ANSWER")
print(ask_rag(question))

print("\n" + "="*80)

print("NAIVE ANSWER")
print(ask_naive(question))
```
